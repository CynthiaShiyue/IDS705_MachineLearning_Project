```python
import pandas as pd

filter_station=pd.read_csv("usfa-registry-CA.csv") #https://apps.usfa.fema.gov/registry/download

filter_station.columns = filter_station.columns.str.strip().str.replace("'", "")
filter_station['HQ zip'] = filter_station['HQ zip'].astype(str).str[:5]
filter_station.columns
```

```python
filter_station[filter_station['HQ state'] == 'CA'].shape[0]
```

```python
filter_station=filter_station[['FDID','Fire dept name','HQ zip','Number Of Stations']]

zip_counts = filter_station.groupby('HQ zip').size().reset_index(name='fire_dept_count')

zip_counts = zip_counts.rename(columns={'HQ zip': 'ZIP_CODE'})

zip_counts.to_csv('zip_counts.csv', index=False)

print("save zip_counts.csv")

```
